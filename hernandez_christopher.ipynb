```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
data = {
    "feature-01": [21, 22, 35, 80, 443],
    "feature-02": [10100, 4000, 4500, 3200, 2900]
}
df = pd.DataFrame(data)
df
```

```python
df["feature-03"] = [random.choice([0,1]) for _ in range(len(df))]
df
```

```python
df.index = ['c1', 'c2', 'c3', 'c4', 'c5']
df
```

```python
plt.plot(df.index, df["feature-01"], marker='o')
plt.title("Feature-01 Graph")
plt.xlabel("Index")
plt.ylabel("Feature-01 Values")
plt.show()
```

```python
plt.bar(df.index, df["feature-03"])
plt.title("Feature-03 Graph")
plt.xlabel("Index")
plt.ylabel("Feature-03 Values")
plt.show()
```
